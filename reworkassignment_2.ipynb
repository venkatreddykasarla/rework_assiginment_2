```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import wbgapi as wb
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Environment indicators
indicator_ids = ['EG.ELC.ACCS.ZS','EG.ELC.NUCL.ZS','EG.ELC.FOSL.ZS']

country_codes = ['IDN','IND','CHN','JPN','NZL','THA','GBR','IRL','CAN','USA']
```

```python
wb.series.info(indicator_ids)
```

```python
# Transpose Function
def readData(file_name,country_codes):
    file = pd.DataFrame(wb.data.DataFrame(file_name, country_codes, time=range(2000, 2016)))
    file_tr = file.transpose()
    return file, file_tr
```

```python
electricity_acc, ele_acc_trans = readData('EG.ELC.ACCS.ZS',country_codes)
electricity_ogs, ele_ogs_trans = readData('EG.ELC.FOSL.ZS',country_codes)
electricity_nuc, ele_nuc_trans = readData('EG.ELC.NUCL.ZS',country_codes)
```

```python
mean_gbr = ele_ogs_trans['GBR'].mean()
median_gbr = ele_ogs_trans['GBR'].median()
strdev_gbr = ele_ogs_trans['GBR'].std()
```

```python
print(' mean of GRATEBRITIAN Electricity production from oil, gas and coal sources',mean_gbr)
print('median of GRATEBRITIAN Electricity production from oil, gas and coal sources',median_gbr)
print('standard deviation of GRATEBRITIAN Electricity production from oil, gas and coal sources',median_gbr)
```

```python
mean_ele = {}
var_ele = {}
median_ele = {}
for i in ele_ogs_trans:
    mean_ele[i] = ele_ogs_trans[i].mean()
    var_ele[i] = ele_ogs_trans[i].var()
    median_ele[i] = ele_ogs_trans[i].median()
    
print('total mean Electricity production from oil, gas and coal sources\n',mean_ele)
print('total variance Electricity production from oil, gas and coal sources\n',var_ele)
print('total median for Electricity production from oil, gas and coal sources\n',median_ele)
```
